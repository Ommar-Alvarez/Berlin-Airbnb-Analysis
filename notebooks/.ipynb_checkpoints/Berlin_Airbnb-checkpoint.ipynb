```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings("ignore")
```

### Creating DF I: Reviews

```python
DATA_DIR = '../../BerlinAirbnbAnalysis/data'
FILE_NAME = 'reviews_summary.csv'
data_path = os.path.join(DATA_DIR, FILE_NAME)
reviews = pd.read_csv(data_path)
reviews.head()
```

```python
# drop unnecessary columns
reviews.drop(['reviewer_id', 'reviewer_name', 'comments'], axis=1, inplace=True)

# rename id column and replace current value simply with 1 to later count the # of reviews per accommodation
reviews.rename({'id':'review_count'}, axis=1, inplace=True)
reviews.review_count = 1

# checking size and duplicates
print("The first dataset has {} rows and {} columns.".format(*reviews.shape))
print("It contains {} duplicates.".format(reviews.duplicated().sum()))

reviews.head()
```

```python
# cast date-column from current object datatype to datetime datatype
reviews.date = pd.to_datetime(reviews.date, format="%Y-%m-%d")
```

```python
review_cnt = reviews[reviews["date"].isin(pd.date_range('2017-11-01', '2018-10-31'))]
print("The dataset 2017/2018 has {} reviews and {} columns.".format(*review_cnt.shape))
```

```python
# set the date as index and sort by index
review_cnt = review_cnt.set_index('date').sort_index()
```

```python
# group by month and listing_id, i.e. specific apartment 
# agg argument: #work on this col:  {#name the new col : #perform count operation}                        
review_cnt = review_cnt.groupby([pd.Grouper(freq='M'), 'listing_id'])['review_count']\
                                         .agg({'review_count':{'reviews_per_month':'count'}})\
                                         .reset_index()
review_cnt.head()
```

```python
# reset index
review_cnt.columns = ['date', 'listing_id', 'reviews_per_month']
review_cnt.set_index('date', inplace=True)
review_cnt = review_cnt.reset_index()
review_cnt.head()
```

```python
# group by listing_id and count reviews, turn it into a dataframe without groupby index
review_cnt = pd.DataFrame(
    review_cnt.groupby('listing_id')['reviews_per_month'].mean().reset_index()
)
review_cnt.head()
```

### Creating DF II: Merging Reviews & Listings

```python
FILE_NAME = 'listings_summary.csv'
data_path = os.path.join(DATA_DIR, FILE_NAME)
listings = pd.read_csv(data_path)
# listings.columns
listings.columns
```

```python
# merge
airbnb = pd.merge(
    review_cnt, listings[['id', 'price', 'cleaning_fee', 'security_deposit', 'neighbourhood_group_cleansed', 
                          'latitude', 'longitude', 'property_type', 'accommodates', 'bathrooms', 'bedrooms',
                          'host_is_superhost', 'instant_bookable', 'review_scores_value', 'amenities', 'bed_type',
                          'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
                          'review_scores_checkin', 'review_scores_communication', 'review_scores_location'
                         ]],
    left_on='listing_id', right_on='id')

# drop 'id' from right dataframe as it's the same as 'listing_id' on the left dataframe
airbnb.drop(['id'], axis=1, inplace=True)

# check the dataframe
print("The dataset has {} rows and {} columns - after combining it.".format(*airbnb.shape))
print("It contains {} duplicates.".format(reviews.duplicated().sum()))
airbnb.columns
```

### Preparing Data for Analaysis I : Calculating Geodesic Distance (from tourist landmarks)
***

```python
from geopy.distance import great_circle

# Calculate geosic distance from Berlin to neighboring neighborhood groups
def distance_to_mid(lat, lon):
    berlin_centre = (52.516667, 13.383333)
    accommodation = (lat, lon)
    return great_circle(berlin_centre, accommodation).km

def distance_to_hbf(lat, lon):
    berlin_centre = (52.52493, 13.369181)
    accommodation = (lat, lon)
    return great_circle(berlin_centre, accommodation).km

def distance_to_txl(lat, lon):
    berlin_centre = (52.559722, 13.287778)
    accommodation = (lat, lon)
    return great_circle(berlin_centre, accommodation).km

def distance_to_museum(lat, lon):
    berlin_centre = (52.521389, 13.395556)
    accommodation = (lat, lon)
    return great_circle(berlin_centre, accommodation).km

def distance_to_reichstag(lat, lon):
    berlin_centre = (52.5186, 13.3763)
    accommodation = (lat, lon)
    return great_circle(berlin_centre, accommodation).km

def distance_to_brandenburg(lat, lon):
    berlin_centre = (52.516272, 13.377722)
    accommodation = (lat, lon)
    return great_circle(berlin_centre, accommodation).km

airbnb['distance_mid'] = airbnb.apply(lambda x: distance_to_mid(x.latitude, x.longitude), axis=1)
airbnb['distance_hbf'] = airbnb.apply(lambda x: distance_to_hbf(x.latitude, x.longitude), axis=1)
airbnb['distance_txl'] = airbnb.apply(lambda x: distance_to_txl(x.latitude, x.longitude), axis=1)
airbnb['distance_museum'] = airbnb.apply(lambda x: distance_to_museum(x.latitude, x.longitude), axis=1)
airbnb['distance_reichstag'] = airbnb.apply(lambda x: distance_to_reichstag(x.latitude, x.longitude), axis=1)
airbnb['distance_brandenburg'] = airbnb.apply(lambda x: distance_to_brandenburg(x.latitude, x.longitude), axis=1)
```

### Preparing Data for Analysis II : Encoding
***

```python
from collections import Counter

# determine which amenities are the most common/important
results = Counter()
airbnb['amenities'].str.strip('{}')\
               .str.replace('"', '')\
               .str.lstrip('\"')\
               .str.rstrip('\"')\
               .str.split(',')\
               .apply(results.update)

results.most_common(30)
```

```python
plt.style.use('seaborn')

# determine which sub-amenities are the most common/important
sub_amenities = pd.DataFrame(results.most_common(60), columns=['amenity', 'count'])
sub_amenities.sort_values(by=['count'],ascending=True).plot(
    kind='barh',
    x='amenity',
    y='count',
    figsize=(10,10),
    legend=False,
    color='black',
    title='Amenities')

plt.xlabel('Count');
```

```python
# amenities that intuitively influence listing price
airbnb['Stove'] = airbnb.amenities.str.contains('Stove')
airbnb['Refrigerator'] = airbnb.amenities.str.contains('Refrigerator')
airbnb['Cooking_basics'] = airbnb.amenities.str.contains('Cooking basics')
airbnb['Long_term_stays_allowed'] = airbnb.amenities.str.contains('Long term stays allowed')
airbnb['Smoking_allowed'] = airbnb.amenities.str.contains('Smoking allowed')
airbnb['Pets_allowed'] = airbnb.amenities.str.contains('Pets allowed')
airbnb['Patio_Balcony'] = airbnb.amenities.str.contains('Patio or balcony')
airbnb['TV'] = airbnb.amenities.str.contains('Patio or balcony')


airbnb.drop(['amenities'], axis=1, inplace=True)
```

```python
# One-hot encoding amenities/sub-ammenities w/high correlation to price

# one_hot_list = ['Host_greets_you', 'Cooking_basics', 'Stove', 'Refrigerator', 'Cooking_basics',
#                 'Long_term_stays_allowed', 'Smoking_allowed', 'Pets_allowed', 'Patio_Balcony']

#test
one_hot_list = ['Cooking_basics', 'Stove', 'Refrigerator', 'Cooking_basics',
                'Long_term_stays_allowed', 'Smoking_allowed', 'Pets_allowed', 'Patio_Balcony', 'TV']

one_hot_col = list(airbnb[one_hot_list])

# Concat encoded variables back to airbnb DF

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Stove, prefix='Stove', drop_first=True)], axis=1)
airbnb.rename(columns={'Stove_True': 'Stove_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Refrigerator, prefix='Refrigerator', drop_first=True)], axis=1)
airbnb.rename(columns={'Refrigerator_True': 'Refrigerator_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Cooking_basics, prefix='Cooking_basics', drop_first=True)], axis=1)
airbnb.rename(columns={'Cooking_basics_True': 'Cooking_basics_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Long_term_stays_allowed, prefix='Long_term_stays_allowed', drop_first=True)], axis=1)
airbnb.rename(columns={'Long_term_stays_allowed_True': 'Long_term_stays_allowed_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Smoking_allowed, prefix='Smoking_allowed', drop_first=True)], axis=1)
airbnb.rename(columns={'Smoking_allowed_True': 'Smoking_allowed_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Pets_allowed, prefix='Pets_allowed', drop_first=True)], axis=1)
airbnb.rename(columns={'Pets_allowed_True': 'Pets_allowed_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.Patio_Balcony, prefix='Patio_Balcony', drop_first=True)], axis=1)
airbnb.rename(columns={'Patio_Balcony_True': 'Patio_Balcony_code'}, inplace='True')

airbnb = pd.concat([airbnb, pd.get_dummies(airbnb.TV, prefix='TV', drop_first=True)], axis=1)
airbnb.rename(columns={'TV_True': 'TV_code'}, inplace='True')


airbnb = airbnb.drop(one_hot_col, axis=1) # drops original binary T/F variables
airbnb = airbnb.loc[:, ~airbnb.columns.duplicated()] # prevents needless concats
airbnb.head()
```

### Preparing Data for Analysis IV: Feature Selection
***

```python
# Selecting features w/explanatory power for price
numerical_features = [
    'reviews_per_month',
    'cleaning_fee',
    'security_deposit',
    'accommodates',
    'bathrooms',
    'bedrooms',
    'review_scores_value',
    'review_scores_rating',
    'review_scores_accuracy',
    'review_scores_cleanliness',
    'review_scores_checkin',
    'review_scores_communication',
    'review_scores_location',
    'distance_mid',
    'distance_hbf',
    'distance_txl',
    'distance_museum',
    'distance_reichstag',
    'distance_brandenburg',
    'Stove_code',
    'Refrigerator_code',
    'Cooking_basics_code',
    'Long_term_stays_allowed_code',
    'Smoking_allowed_code',
    'Pets_allowed_code',
    'Patio_Balcony_code',
    'Apartment', 'Condomimium', 'Loft', 'House',
    'Charlottenburg-Wilm.','Friedrichshain-Kreuzberg', 'Lichtenberg', 'Marzahn-Hellersdorf','Mitte', 'Neukölln', 'Pankow', 'Reinickendorf', 'Spandau', 'Steglitz-Zehlendorf', 'Tempelhof-Schöneberg', 'Treptow-Köpenick',
    'airbed_bed', 'couch_bed', 'futon_bed', 'pull_out_bed', 'real_bed',
    'instant_bookable_code',
    'superhost_code',
    'TV_code'
]

# Feature selection based on correlation to price 
print('Corr(price, βs)')
airbnb[numerical_features].corrwith(airbnb.price).sort_values(ascending=False).round(3)

```

```python
from sklearn.model_selection import train_test_split

X = airbnb[numerical_features]
y = airbnb.price

# train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=123)
```

### Model Selection I: Unclustered
***

#### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

# fitting
l_reg = LinearRegression()
l_reg.fit(X_train, y_train)
y_pred = l_reg.predict(X_test)

# coefficient estimates
print('Linear Regression coefficients')
pd.Series(l_reg.coef_, index=X_train.columns).sort_values(ascending=False).round(2)
```

```python
import statsmodels.api as sm

# Only return parametrs that are statistically significant
results = sm.OLS(y_train, X_train).fit()
round(results.params[results.pvalues < 0.05], 3).sort_values(ascending=False)
```

#### Linear Regression Notes
- Smoking_allowed_code may be negative due to customers (on average) prefering listings that w/o the smell of smoke
- review_scores_value is negative b/c low price impacts expected value
    - e.g. If a customer had a great experience for a low price, value would increase
- distance_txl may be closer to Berlin center than distance_brandenburg, hence polarities

```python
from sklearn.metrics import r2_score, mean_squared_error

print('Linear Regression Performance')
print()
print('MSE score: %.2f' % mean_squared_error(y_test, y_pred))
print('R2 score: %.2f' % r2_score(y_test, y_pred))
```

#### Lasso Regression Model

```python
from sklearn.linear_model import Lasso

lasso = Lasso(
    alpha=0, # α → 0 = LR; α → ∞ = coeffs equal to 0
)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)

pd.Series(lasso.coef_, index=X_train.columns).sort_values(ascending=False).round(2)
```

```python
print('Lasso Regression Performance')
print()
print('MSE score: %.2f' % mean_squared_error(y_test, y_pred))
print('R2 score: %.2f' % r2_score(y_test, y_pred))
```

#### KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(
    n_neighbors=10,
    weights='distance',
)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print('KNN Performance')
print()
print('MSE score: %.2f' % mean_squared_error(y_test, y_pred))
print('R2 score: %.2f' % r2_score(y_test, y_pred))
```

#### KNN Regression Notes
- outperformed by linear regression

### Model Selection II: Cluster Models
